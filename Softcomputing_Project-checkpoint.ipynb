```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Medical Cost Personal Datasets.csv")

# Age Distribution Graph

plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=10, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Charges by Age

plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='charges', data=df)
plt.title('Medical Charges by Age')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()


# BMI Distribution by Smoking Status

plt.figure(figsize=(10, 6))
sns.boxplot(x='smoker', y='bmi', data=df)
plt.title('BMI Distribution by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('BMI')
plt.show()


# Medical Charges by Region

plt.figure(figsize=(10, 6))
sns.barplot(x='region', y='charges', data=df, estimator=sum)
plt.title('Total Medical Charges by Region')
plt.xlabel('Region')
plt.ylabel('Total Charges')
plt.show()


# Correlation Heatmap

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Medical Charges by Number of Children

plt.figure(figsize=(10, 6))
sns.barplot(x='children', y='charges', data=df)
plt.title('Medical Charges by Number of Children')
plt.xlabel('Number of Children')
plt.ylabel('Charges')
plt.show()


# Medical Charges Distribution by Gender

plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='charges', data=df)
plt.title('Medical Charges Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Charges')
plt.show()

# Pairplot for Relationship Analysis

sns.pairplot(df[['age', 'bmi', 'charges', 'children']], hue='smoker', diag_kind='kde')
plt.show()
```
